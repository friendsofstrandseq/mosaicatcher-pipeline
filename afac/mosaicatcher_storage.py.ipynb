```python
import pandas as pd 
# import plotly.express as px

data = pd.read_csv("/g/korbel/STOCKS_WF/mosaicatcher_pipeline_summary.tsv", sep="\t").reset_index()
data.columns = ["flowcell", "sample", "folder", "size", "_", "size%"]
data = data.drop(columns="_")
data = data.dropna()
data
```

```python
import numpy as np
def convert_size_to_bytes(size):
    if pd.isna(size):  # Check if the value is NaN
        return np.nan
    size = str(size)  # Convert the value to a string first
    if "K" in size:
        return int(float(size[:-1]) * 1024)
    elif "M" in size:
        return int(float(size[:-1]) * 1024 * 1024)
    elif "G" in size:
        return int(float(size[:-1]) * 1024 * 1024 * 1024)
    else:
        return int(float(size))  # Assume the value is already in bytes

data["size"] = data["size"].apply(convert_size_to_bytes)

data["size"] = data["size"].apply(lambda x: x/1024/1024/1024) # Convert bytes to GB
data["size%"] = data["size%"].apply(lambda x: float(str(x)[:-1])) # Remove the % sign
data
# data["size"].sum()
```
